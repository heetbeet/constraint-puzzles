```python
import urllib
import urllib.request
import os
import time

import urllib
import urllib.request

def get_page(url):
    return urllib.request.urlopen(url).read().decode("iso-8859-1")

def is_int(txt):
    try:
        int(txt)
        return True
    except:
        return False
    
chlist = []
diffs = 'IM CH TF ST IN'.split()
diffsfull = 'intermediate challenging tough super-tough insane'.split()
for vol in range(1,11):
    for bk in range(1,101):
        for pz in range(1,9):
            for ch in diffs:
                chlist.append([vol,bk,pz,ch])
```

```python
import numpy as np
from itertools import product

while(True):
    try:
        completion_dict = {}
        for dir, dirs, files in os.walk('../killersudoku-krazydad-collection/'):
            for file in files:
                if len(open(dir+'/'+file, 'r').read())>100:
                    completion_dict[int(dir.split('diff')[-1]),
                                    int(file.split('vol')[1].split('-')[0]),
                                    int(file.split('book')[1].split('-')[0]),
                                    int(file.split('puzzle')[1].split('.')[0])] = True

        for vol, bk, pz, ch in chlist:
            chidx = diffs.index(ch)+1
            if (chidx, vol, bk, pz) in completion_dict:
                continue

            url = 'https://krazydad.com/tablet/killer/?kind=%s&volumeNumber=%d&bookNumber=%d&puzzleNumber=%d'%(ch,vol,bk,pz)
            str_in = get_page(url)
            if 'Puzzle not found' in str_in: 
                print('no puzzle for', vol, bk, pz, ch)
                continue
            else:
                print(vol, bk, pz, ch)


            for line in str_in.split('\n'):
                if '"puzz"' in line:
                    cages, sums = line.split('"puzz"')[1].split('"')[1][1:].split(';')

                    cages = np.array([i for i in cages]).reshape(9,-1)
                    sums = [int(i.strip()) for i in sums.split(',')]
                    break

            cagedict = {}
            for i,j in product(range(9), range(9)):
                try:
                    cagedict[cages[i,j]].append((i,j))
                except:
                    cagedict[cages[i,j]] = [(i,j)]

            str2num = {j:i for i,j in enumerate('ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz')}
            cells = {}
            for key, val in cagedict.items():
                cells[sums[str2num[key]], val[0]] = val

            txt = '#source: '+url+'\n'
            txt += '#reported difficulty: '+str(diffsfull[chidx-1])+'\n'
            txt += '#creation_date: '+str_in.split('"creation_date": "')[1].split('"')[0]+'\n\n'
            txt += '+'+('---+'*9)+'\n'
            for i in range(0,9):
                txt += '|'
                markers = []
                for j in range(0,9):
                    isin = False
                    for (sm, fval), vals in cells.items():
                        if (i,j) in vals:
                            break

                    txt += (str(sm)+'   ')[:3] if fval == (i,j) else '   '
                    isthick = (j+1)%3==0 and j<8 
                    txt += ('|║')[isthick] if (i,j+1) not in vals else '  '[isthick]
                    markers.append(not (i+1,j) in vals)
                txt += '\n'
                isthick = (i+1)%3==0 and i<8
                txt += '+' + '+'.join(('---','===')[isthick] if i else ('   ','   ')[isthick] for i in markers) + '+\n'


            filename = '../killersudoku-krazydad-collection/diff%02d/vol%02d-book%03d-puzzle%02d.txt'%(chidx, vol, bk, pz)
            os.makedirs(os.path.dirname(filename), exist_ok=True)
            open(filename, 'w').write(txt)
        break
    except Exception as e:
        print('timeout...', e)
        time.sleep(5)
```

```python
#--------------------------------------------------
# convert to emdw readables for emdw side of things
#--------------------------------------------------

import sys
import os
if not '..' in sys.path: sys.path.append("..")
import killersudoku

dirs = set()
for dir, _, files in os.walk('../killersudoku-krazydad-collection/'):
    for file in files:
        file = dir+'/'+file
        cages = killersudoku.load_killersudoku(file)
        fileout = file.replace('collection', 'collection--emdw').replace('.txt', '.ksd')
        os.makedirs(os.path.dirname(fileout), exist_ok=True)
        
        str_out = ''
        for cage in cages:
            str_out += str(cage[0]) +' ' + ' '.join( str(i)+str(j) for i,j in cage[1]) + '\n'

        open(fileout, 'w').write(
            str_out
        )
```
